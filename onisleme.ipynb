```python
import re
```

```python
def duzelt(wikitext):
  wikitext = re.sub(':\[1\]', '#', wikitext)
  wikitext = re.sub(':\[2\]', '#', wikitext)
  wikitext = re.sub(':\[3\]', '#', wikitext)
  wikitext = re.sub(':\[4\]', '#', wikitext)
  wikitext = re.sub(':\[5\]', '#', wikitext)
  return wikitext
```

## Parse

```python
!pip install wikitextparser
import wikitextparser as wtp
```

```python
def ayikla(wikitext):
  wikitext = duzelt(wikitext)
  sections = wtp.parse(wikitext).get_sections(include_subsections=False, level=3) # Köken & Atasözü
  for s in sections:
    if s.title == 'Atasözü':
      return s.get_lists(pattern='\#')[0].plain_text().split('\n') # plain_text() - remove_markups()
  return "" # "Atasözü" bölümünü bulamazsa
```

```python
def cikar(wikitext):
  try:
    maddeler = ayikla(wikitext)
  except IndexError: # "Atasözü" bölümünü buldu fakat içerisinde madde yok
    return ""
  else:
    metin = ' '.join(maddeler)
    metin = re.sub('[\[\]1-9]', '', metin) # hyperlink işaretleri [[kelime]] ve rakamlar 1-9
    return metin
```

```python
df['anlam'] = df['wikitext'].apply(cikar)
```

## Fuzzy Duplicates
- https://towardsdatascience.com/fuzzywuzzy-find-similar-strings-within-one-column-in-a-pandas-data-frame-99f6c2a0c212

```python
# https://github.com/seatgeek/thefuzz
!pip install thefuzz[speedup]
from thefuzz import process
from thefuzz import fuzz
```

```python
# bu veri kümesinde yaptığım testlere göre; QRatio > WRatio (tabii bu veri kümesi için)
# UQRatio = QRatio with Unicode : Karakterler Türkçe olduğu için
df['muhtemel aynılar'] = ""
for i in range(len(df)):
  asz = df.iloc[i,1]
  altindakiler = df.iloc[i+1:, 1]
  eslesenler = [t for t in process.extractBests(asz, altindakiler, scorer=fuzz.UQRatio, score_cutoff=75)]
  for e in eslesenler:
    df.iat[i,4] = e
```
